```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_study_df = pd.merge (study_results, mouse_metadata, on= "Mouse ID", how= "left")

# Display the data table for preview
mouse_study_df.head()
```

```python
# Checking the number of mice.mice_number = len(mouse_study_df["Mouse ID"].unique())
mice_number = len(mouse_study_df['Mouse ID'].unique())
mice_number
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
#https://www.geeksforgeeks.org/find-duplicate-rows-in-a-dataframe-based-on-all-or-selected-columns/
duplicated= mouse_study_df.loc[mouse_study_df.duplicated(["Mouse ID", "Timepoint"])]

duplicated_mouse = duplicated["Mouse ID"].unique()
duplicated_mouse
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
g989_data = mouse_study_df.loc[mouse_study_df['Mouse ID']== 'g989' ]
g989_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
new_mice_df = mouse_study_df.drop(mouse_study_df[mouse_study_df['Mouse ID'] == "g989"].index)
new_mice_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
new_mice_number = len(new_mice_df['Mouse ID'].unique())
new_mice_number
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
tumor_volume_mean = new_mice_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
tumor_volume_median = new_mice_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
tumor_volume_variance = new_mice_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
tumor_volume_stdv = new_mice_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
tumor_volume_sem = new_mice_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
# Assemble the resulting series into a single summary DataFrame.
summary_df = pd.DataFrame({
    "Mean Tumor Volume" : tumor_volume_mean,
    "Median Tumor Volume" : tumor_volume_median,
    "Tumor Volume Variance" : tumor_volume_variance,
    "Tumor Volume Std.Dev." : tumor_volume_stdv,
    "Tumor Volume Std.Err." : tumor_volume_sem
})
summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_statistics = new_mice_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_statistics
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_tp_df = new_mice_df.groupby('Mouse ID')['Timepoint'].max().reset_index()
#max_tp_df

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
tumor_at_maxtp = pd.merge(max_tp_df, new_mice_df, on=['Mouse ID', 'Timepoint'], how= "left")
tumor_at_maxtp.head(10)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimens =["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_cap = []
tumor_volume_ram = []
tumor_volume_inf = []
tumor_volume_cef = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes

for i in tumor_at_maxtp.index:
    if tumor_at_maxtp.iloc[i]['Drug Regimen'] == "Capomulin":
           tumor_volume_cap.append(tumor_at_maxtp.iloc[i]['Tumor Volume (mm3)'])
    if tumor_at_maxtp.iloc[i]['Drug Regimen'] == "Ramicane":
           tumor_volume_ram.append(tumor_at_maxtp.iloc[i]['Tumor Volume (mm3)'])
    if tumor_at_maxtp.iloc[i]['Drug Regimen'] == "Infubinol":
           tumor_volume_inf.append(tumor_at_maxtp.iloc[i]['Tumor Volume (mm3)'])
    if tumor_at_maxtp.iloc[i]['Drug Regimen'] == "Ceftamin":
           tumor_volume_cef.append(tumor_at_maxtp.iloc[i]['Tumor Volume (mm3)'])
     
      
     # add subset 
     
tumor_vol_df = pd.DataFrame({"Capomulin" : tumor_volume_cap,
                              "Ramicane" : tumor_volume_ram,
                              "Infubinol" : tumor_volume_inf,
                              "Ceftamin" : tumor_volume_cef})



    
     # Determine outliers using upper and lower bounds
for regimen in regimens:    
    tumor_vol = tumor_vol_df[regimen]
    quartiles = tumor_vol.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq- lowerq

    lower_limit = lowerq - (1.5 * iqr)
    upper_limit = upperq + (1.5 * iqr)

    #outliers = tumor_vol_df.loc[(tumor_vol_df[regimen] < lower_limit) | (tumor_vol_df[regimen] > upper_limit) ]
    outliers = tumor_vol.loc[(tumor_vol < lower_limit) | (tumor_vol > upper_limit) ]
    print(f"{regimen}'s potential outliers: {outliers}")


```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

tumor_vol_data =[(tumor_vol_df['Capomulin']),
                 (tumor_vol_df['Ramicane']), 
                 (tumor_vol_df['Infubinol']),
                 (tumor_vol_df['Ceftamin'])]
fig1, ax = plt.subplots(figsize= (6.5 ,5) )
ax.set_ylabel("Final Tumor Volume(mm3)")
ax.set_xlabel("Drug Regimens")
ax.set_title("Final tumor Volume by Drug Regimen")
#https://matplotlib.org/stable/gallery/statistics/boxplot.html#sphx-glr-gallery-statistics-boxplot-py
flierprops = dict(marker="o", markerfacecolor="red", markersize= 10, markeredgecolor= "black")
ax.boxplot(tumor_vol_data,flierprops= flierprops)
plt.xticks([1,2,3,4], regimens)
plt.show
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
mouse_l509 = new_mice_df.loc[new_mice_df['Mouse ID']== 'l509']
#mouse_l509
x_axis = mouse_l509["Timepoint"]
y_axis = mouse_l509['Tumor Volume (mm3)']
plt.plot(x_axis, y_axis)
plt.title("Capomulin treatment of mouse l509")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
cap_df = new_mice_df.loc[new_mice_df['Drug Regimen'] == "Capomulin", :]
#cap_avg = cap_df.groupby(['Mouse ID'])[].mean()
avg_mass_cap = cap_df.groupby(['Mouse ID'])['Weight (g)'].mean()
avg_vol_cap = cap_df.groupby(['Mouse ID'])['Tumor Volume (mm3)'].mean()

plt.scatter(avg_mass_cap, avg_vol_cap)
plt.title("Mouse Weight vs average Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(avg_mass_cap, avg_vol_cap)
print(f"The correlation between mouse weight and the average tumor volume is{round(correlation[0],2)}")

(slope, intercept, rvalue, pvalue, stderr)= st.linregress(avg_mass_cap, avg_vol_cap)
regress_values = avg_mass_cap * slope + intercept
line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept,2))

plt.scatter(avg_mass_cap, avg_vol_cap)
plt.plot(avg_mass_cap,regress_values, "r-")
plt.annotate(line_eq, (20,35), fontsize= 14, color= "r")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
